```python
import random
class TicTacToe:
  def __init__(self):
    self.counter = 0
    self.createboard()

  def createboard(self):
    board = [[], [], []]
    for i in range(3):
      for j in range(3):
        board[i].append('-')
    self.board = board

  def startgame(self):
    self.first = random.choice('XO')
    self.second = 'X' if self.first =='O' else 'O'
    self.playerturn(self.first, self.second)

  def playerturn(self, p1, p2):
    while True:
      self.counter += 1
      check = all('-' not in row for row in self.board)
      if check:
        self.showboard()
        print("Game Over")
        break
      elif self.counter % 2 != 0 :
        win = self.player1(p1)
        if win:
          self.showboard()
          print('Player', p1, 'wins')
          break
      else:
        win = self.player2(p2)
        if win:
          self.showboard()
          print('Player', p1, 'wins')
          break

  def player1(self, p1):
    print('\n')
    print('Player', p1, 'turn')
    self.showboard()
    print('\n')
    a,b = input(print('Enter the row and column number you want to pick'))
    self.fillpoint(a,b,p1)
    if self.counter >= 5:
      win = self.checkwin(p1)
      return win

  def player2(self, p2):
    print('\n')
    print('Player', p2, 'turn')
    self.showboard()
    print('\n')
    a,b = input(print('Enter the row and column number you want to pick'))
    self.fillpoint(a,b,p2)
    if self.counter > 5:
      win = self.checkwin(p2)
      return win

  def showboard(self):
    for row in self.board:
      for i in row:
        print(i, ' ' , end = '')
      print('\n')

  def fillpoint(self,a,b, player):
    if self.board[int(a)][int(b)] == '-':
      self.board[int(a)][int(b)] = player
    else:
      print('\n')
      print('This field is already filled up, Please choose another one')
      self.player1(player) if player == self.first else self.player2(player)

  def checkwin(self, p):
    win = False
    if self.board[0][1] == self.board[0][1] == self.board[0][2] == p:
      win = True
    elif self.board[1][0] == self.board[1][1] == self.board[1][2] == p:
      win = True
    elif self.board[2][0] == self.board[2][1] == self.board[2][2] == p:
      win = True
    elif self.board[0][0] == self.board[1][0] == self.board[2][0] == p:
      win = True
    elif self.board[0][1] == self.board[1][1] == self.board[2][1] == p:
      win = True
    elif self.board[0][2] == self.board[1][2] == self.board[2][2] == p:
      win = True
    elif self.board[0][0] == self.board[1][1] == self.board[2][2] == p:
      win = True
    elif self.board[0][2] == self.board[1][1] == self.board[2][0] == p:
      win = True
    return win



```
